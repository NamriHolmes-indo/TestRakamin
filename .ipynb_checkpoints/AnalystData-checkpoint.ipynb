```python
query_profiles_psych = "SELECT * FROM testrakamin.profiles_psych"
query_strengths = "SELECT * FROM testrakamin.strengths"
query_papi_scores = "SELECT * FROM testrakamin.papi_scores"

data_profiles_psych = pd.read_sql(query_profiles_psych, engine)
data_strengths = pd.read_sql(query_strengths, engine)
data_papi_scores = pd.read_sql(query_papi_scores, engine)
```

```python
def show_table_summary(df, name):
    print(f"\n==================== {name} ====================")
    print("INFO:")
    print(df.info())
    print("\nDESCRIBE:")
    print(df.describe(include='all'))
    
    print("\nVALUES COUNT PER KOLOM:")
    df_noid = df.drop("employee_id", axis=1)
    for col in df_noid.columns:
        print(f"\n--- {col} ---")
        print(df_noid[col].value_counts(dropna=False).head(10))

show_table_summary(data_profiles_psych, "PROFILES_PSYCH")
show_table_summary(data_strengths, "STRENGTHS")
show_table_summary(data_papi_scores, "PAPI_SCORES")        
```

```python
from reportlab.lib.pagesizes import A4
from reportlab.lib.styles import getSampleStyleSheet
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer
from reportlab.lib.units import cm
from io import StringIO

def get_summary_text(df, name):
    buffer = StringIO()
    buffer.write(f"\n==================== {name} ====================\n")
    buffer.write("INFO:\n")
    buffer.write(str(df.info(buf=StringIO())) + "\n")
    buffer.write("\nDESCRIBE:\n")
    buffer.write(str(df.describe(include='all')) + "\n")
    buffer.write("\nVALUES COUNT PER KOLOM:\n")
    df_noid = df.drop("employee_id", axis=1, errors="ignore")
    for col in df_noid.columns:
        buffer.write(f"\n--- {col} ---\n")
        buffer.write(str(df_noid[col].value_counts(dropna=False).head(10)) + "\n")
    return buffer.getvalue()

text_profiles = get_summary_text(data_profiles_psych, "PROFILES_PSYCH")
text_strengths = get_summary_text(data_strengths, "STRENGTHS")
text_papi = get_summary_text(data_papi_scores, "PAPI_SCORES")

full_text = text_profiles + "\n\n" + text_strengths + "\n\n" + text_papi

doc = SimpleDocTemplate("summary_report.pdf", pagesize=A4)
styles = getSampleStyleSheet()
story = []

for section in full_text.split("\n\n"):
    story.append(Paragraph(section.replace("\n", "<br/>"), styles["Normal"]))
    story.append(Spacer(1, 0.5*cm))

doc.build(story)
print("✅ Laporan berhasil diexport ke summary_report.pdf")
```

```python
from reportlab.lib.pagesizes import A4
from reportlab.lib import colors
from reportlab.platypus import SimpleDocTemplate, Paragraph, Spacer, Table, TableStyle, PageBreak
from reportlab.lib.styles import getSampleStyleSheet

def dataframe_to_table(df, title):
    data = [df.columns.tolist()] + df.astype(str).values.tolist()
    table = Table(data, repeatRows=1)
    style = TableStyle([
        ('BACKGROUND', (0,0), (-1,0), colors.lightblue),
        ('TEXTCOLOR', (0,0), (-1,0), colors.black),
        ('ALIGN', (0,0), (-1,-1), 'CENTER'),
        ('FONTNAME', (0,0), (-1,0), 'Helvetica-Bold'),
        ('BOTTOMPADDING', (0,0), (-1,0), 6),
        ('GRID', (0,0), (-1,-1), 0.25, colors.grey),
        ('FONTNAME', (0,1), (-1,-1), 'Helvetica'),
        ('FONTSIZE', (0,0), (-1,-1), 7),
    ])
    table.setStyle(style)
    return [Paragraph(f"<b>{title}</b>", getSampleStyleSheet()['Heading4']), table, Spacer(1, 8)]

def show_table_summary_to_pdf(df, name, story):
    story.append(Paragraph(f"<b>==================== {name} ====================</b>", getSampleStyleSheet()['Heading3']))
    story.append(Spacer(1, 8))

    info_df = pd.DataFrame({
        "Column": df.columns,
        "Non-Null Count": df.notna().sum(),
        "Null Count": df.isna().sum(),
        "Dtype": df.dtypes.astype(str)
    })
    story += dataframe_to_table(info_df, "INFO")

    desc = df.describe(include='all').transpose().reset_index()
    desc.rename(columns={'index': 'Column'}, inplace=True)
    story += dataframe_to_table(desc, "DESCRIBE")

    df_noid = df.drop("employee_id", axis=1, errors='ignore')
    for col in df_noid.columns:
        vc = df_noid[col].value_counts(dropna=False).reset_index()
        vc.columns = [col, 'count']
        story += dataframe_to_table(vc.head(10), f"TOP 10 VALUES: {col}")

    story.append(PageBreak())

doc = SimpleDocTemplate("summary_report_tabel.pdf", pagesize=A4)
story = []

show_table_summary_to_pdf(data_profiles_psych, "PROFILES_PSYCH", story)
show_table_summary_to_pdf(data_strengths, "STRENGTHS", story)
show_table_summary_to_pdf(data_papi_scores, "PAPI_SCORES", story)

doc.build(story)
print("✅ Laporan berhasil diexport ke summary_report_tabel.pdf dalam format tabel yang rapi.")
```

```python
query_vw_employees_performa = "SELECT * FROM testrakamin.vw_employees_performa"

data_vw_employees_performa = pd.read_sql(query_vw_employees_performa, engine)
print(data_vw_employees_performa.info())
```

```python
drop_cols = ['employee_id', 'fullname', 'nip']
numeric_df = data_vw_employees_performa.drop(columns=drop_cols, errors='ignore').select_dtypes(include='number')

correlation = numeric_df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(
    correlation,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=0.5,
    cbar_kws={"shrink": 0.7}
)
plt.title("Korelasi antar Kolom Numerik - vw_employees_performa", fontsize=14, pad=12)
plt.tight_layout()
plt.show()
```

```python
data_vw_employees_performa = data_vw_employees_performa.sort_values(by=['employee_id', 'year']).reset_index(drop=True)
data_vw_employees_performa['rating_prev'] = data_vw_employees_performa.groupby('employee_id')['rating'].shift(1)
data_vw_employees_performa['delta_rating'] = data_vw_employees_performa['rating'] - data_vw_employees_performa['rating_prev']

data_vw_employees_performa['trend'] = data_vw_employees_performa['delta_rating'].apply(
    lambda x: 'Naik' if x > 0 else ('Turun' if x < 0 else 'Tetap')
)

rating_naik = data_vw_employees_performa[data_vw_employees_performa['trend'] == 'Naik']
print("\n=== Karyawan dengan Rating Naik ===")
print("Karyawan yang rating nya naik:", len(rating_naik))
print(rating_naik[['employee_id', 'fullname', 'year', 'rating_prev', 'rating', 'trend']])
```
